Question Number:4

What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

```python
plt.figure(figsize=(9,4))       
sns.boxplot(y='gender',x='math score',data=data)
plt.ylabel("Gender",fontsize=15)
plt.xlabel("Math score",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(9,4))       
sns.boxplot(y='gender',x='writing score',data=data)
plt.ylabel("Gender",fontsize=15)
plt.xlabel("Writing score",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(9,4))       
sns.boxplot(y='gender',x='reading score',data=data)
plt.ylabel("Gender",fontsize=15)
plt.xlabel("Reading score",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(9,4))       
sns.boxplot(y='test preparation course',x='math score',data=data)
plt.ylabel("Test Preparation Course",fontsize=15)
plt.xlabel("Math score",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(9,4))       
sns.boxplot(y='test preparation course',x='reading score',data=data)
plt.ylabel("Test Preparation Course",fontsize=15)
plt.xlabel("Reading score",fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(9,4))       
sns.boxplot(y='test preparation course',x='writing score',data=data)
plt.ylabel("Test Preparation Course",fontsize=15)
plt.xlabel("Writing score",fontsize=15)
plt.show()
```

From the above boxplots we can see the variation of the math score,reading score and writing score with respect to gender and test preparation course.
